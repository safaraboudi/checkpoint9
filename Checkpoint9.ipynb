```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import mlxtend

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

#Association rules for a clothing company

```python
dataset=[['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
        ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    
data=pd.DataFrame(te_ary, columns=te.columns_)  
data
```

```python
data.columns
```

```python
frequent_itemsets=apriori(data, min_support=0.6, use_colnames=True) 
frequent_itemsets
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=0.7)
```

There is a high probability for customers to buy Skirts and Pants together(lift = 1.25 > 1)so these 2 items should be put together.

#Market Basket Optimisation

```python
df=pd.read_csv('/content/Market_Basket_Optimisation.csv',header=None)
df.head()
```

```python
df.shape
```

```python
df.info()
```

We have to convert the DataFrame into list of lists to do the transaction encoding

```python
lists = []
for i in range (0, 7501):
    lists.append([str(df.values[i,j]) for j in range(0, 20)])
```

```python
TE = TransactionEncoder()
array = TE.fit(lists).transform(lists)
T_df = pd.DataFrame(array, columns = TE.columns_)
T_df
```

We should not forget that the nan value was also encoded so we have to drop the column nan from DataFrame T_df.

```python
T_df.drop(['nan'],axis=1,inplace=True)
```

```python
a_rules = apriori(T_df, min_support = 0.04, use_colnames = True)
a_rules
```

```python
association_rules(a_rules, metric = 'lift', min_threshold = 1)
```

So this means when the customer will buy one of those items: chocolate,eggs,ground beef,milk or spaghetti he will definitly buy mineral water with it.
